# Gasto de Passagens Aéreas realizadas pelo Governo 

```python
# Importando os dados de passagens aéreas
import pandas as pd
```

```python
# Ler o arquivo e criar o DataFrame
df = pd.read_excel('201710_EmissaoPassagens_SCDP.xlsx', skiprows=3, skipfooter=3)
```

```python
# Verificando inicio dos dados
df.head()
```

```python
# Verificando o final dos dados
df.tail()
```

```python
# Verificando dados aleátorios
df.sample(5)
```

```python
# Descobrindo os tipos dos dados
df.dtypes
```

```python
# Retirando os espaços do nome das colunas
df.rename(columns=lambda x: x.strip(), inplace=True)
```

```python
# Verificando se os espaços foram removidos
df.dtypes
```

```python
# Converter a coluna data emissão bilhete para datetime
df['Data Emissão Bilhete'] = pd.to_datetime(df['Data Emissão Bilhete'])
```

```python
# Converter a coluna Data Embarque para datetime
df['Data Embarque'] = pd.to_datetime(df['Data Embarque'])
```

```python
# Verificando se as datas foram convertidas
df.info()
```

```python
# Retirando os espaços do campo Classe Tarifária Bilhete
df['Classe Tarifária Bilhete'] = df['Classe Tarifária Bilhete'].str.strip()
```

```python
# Retirando os espaços do campo No Show
df['No Show'] = df['No Show'].str.strip()

# Retirando os espaços do campo Remarcado
df['Remarcado'] = df['Remarcado'].str.strip()

# Retirando os espaços do campo Cancelado 
df['Cancelado'] = df['Cancelado'].str.strip()
```

```python
# Iniciando as análises
# Verificando quantos orgão do governo utilizam o sistema de Passagen Aéreas e a relação das maiores e menores utilização
# do sistema
df['Nome do órgão superior'].value_counts()
```

```python
# Quantidade de voos cancelados
df['Cancelado'].value_counts()
```

```python
# Quantidade de voos remarcados
df['Remarcado'].value_counts()
```

```python
# Quantidade de voos que não compareceram
df['No Show'].value_counts()
```

```python
# Verificando qual o custo total do governo com passagens aéreas
df['Valor Bilhete'].sum()
```

```python
# Trabalhando com Gráficos
# importando o seaborn
import seaborn as sns
```

```python
# importando o matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df['Data Emissão Bilhete'].dt.day.value_counts().plot.bar()
plt.title('Quantidade de passagens emitidas por dia')
```

```python
# Valor das passagens emitidas por Companhias
sns.violinplot(x='Companhia Aérea', y='Valor Bilhete', data=df)
```
